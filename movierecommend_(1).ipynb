```python
# genres
# keywords
# overview
# title
# movie id
# cast
# crew(only director)
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
#movies.isnull().sum() 3 overviews are empty
movies.dropna(inplace=True)
movies.shape
```

```python
import ast
```

```python
#function will return a list by taking names from the respective field from the dictionary that acts like an object in the list of dictionaries.
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_3(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter< 3:
            L.append(i['name'])
            counter+=1

    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if(i['job']=='Director'):
            L.append(i['name'])

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #converts overview text to list
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x)) #list to string
new_df.head()
```

```python
new_df['tags'][3999]
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)  #finding the cosine distance between vectors
```

# Recommend function

```python
# Here we need to make sure that reverse is True and the tuple is sorted acc to index[1] i.e. the actual distance.
#enumerate similarity converts ndarray into a tuple with index
sorted(list(enumerate(similarity[11])),reverse=True, key=lambda x: x[1])
```

```python
def recommend(search_title):
    index = new_df[new_df['title'] == search_title].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True, key=lambda x: x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].title)
        #print(i)
```

```python
recommend('Prisoners')
```
